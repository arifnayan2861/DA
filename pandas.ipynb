```python
import numpy as np
import pandas as pd
```

# series

```python
#creating series from list
data = [1,2,3,4,5]
series = pd.Series(data)
print(series, type(series))
```

```python
#creating series from dictionary
data2 = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
series2 = pd.Series(data2)
print(series2, type(series2))
```

```python
data3 = ['a', 'b', 'c']
series3 = pd.Series(data3, index=[1,2,3])
print(series3, type(series3))
```

# dataframes

```python
data = {
    'Name': ['John', 'Jane', 'Jack', 'Jill'],
    'Age': [25, 30, 35, 40],
}
df = pd.DataFrame(data)
print(df, type(df))
```

```python
data = [
    {"name": "john", "age": 25},
    {"name": "jane", "age":30},
    {"name": "jack", "age": 35},
    {"name": "jill", "age": 40}
]
df = pd.DataFrame(data)
print(df)
```

```python
data = [
    ["john",25], ["jane", 30], ["jack", 35], ["jill", 40]
]
df = pd.DataFrame(data, columns=["name", "age"])
print(df)
```

```python
df["Sales_fillNA"] = df["Sales"].fillna(df["Sales"].mean())
df
```

```python
df.dtypes
```

```python
#renaming columns
df=df.rename(columns={"Date": "Sales Date"})
df.head()
```

```python
#change data types
df["Value_int"] = df["Value"].fillna(df["Value"].mean()).astype(int)
df
```

```python
df["New Value"] = df["Value_int"].apply(lambda x : x * 2)
df.head()
```

```python
#data aggregating & grouping
df.head()
```

```python
grouped_mean = df.groupby("Product")['Value_int'].mean()
grouped_mean
```

```python
df.groupby("Product")['Value_int'].agg(['mean', 'sum', 'count'])
```

```python
df.groupby(["Product", "Region"])["Value_int"].agg(['count', 'sum', 'mean'])
```
