```python
import matplotlib.pyplot as plot
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_excel("/content/02. flight-price.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking for null value
df.isnull().sum()
```

```python
# splitting the initial date into 3 other column with int data type
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
df.head()
```

```python
df.info()
```

```python
# dropping the initial date column
df.drop('Date_of_Journey', inplace=True, axis=1)
df.head()
```

```python
# removing the extra info from the column
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
# splitting the initial time into two different column
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1].astype(int)
df.head()
```

```python
# dropping the initial arrival_time column
df.drop('Arrival_Time', inplace=True, axis=1)
df.info()
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min']=df['Dep_Time'].str.split(':').str[0].astype(int)
df.drop('Dep_Time', axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Total_Stops'].value_counts()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 1})
df[df['Total_Stops'].isnull()]
```

```python
df.head()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

```python
df[df['Duration'].isnull()]
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
df.drop('Duration', axis=1, inplace=True)
df.head()
```

```python
df[df['Duration_min'].isnull()] = 0
```

```python
df.info()
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
df['Duration_min']=df['Duration_min'].astype(int)
```
